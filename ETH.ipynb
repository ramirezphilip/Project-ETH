#Plotting Ethereum to USD

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importing necessary libraries

```python
ethusd = pd.read_csv('BITFINEX-ETHUSD.csv').set_index('Date')
mid = ethusd.loc[:, 'Mid']
fig, eth = plt.subplots()
eth.plot(ethusd.index, mid)
eth.xaxis.set_major_locator(plt.MaxNLocator(8))
eth.set_ylabel('Value (rounded to nearest USD)')
eth.set_xlabel('Date')
eth.set_title('Ethereum Coin Value over time')
eth.invert_xaxis()
```

Plotting the csv file "BITFINEX-ETHUSD.csv" over its max range

Only using 8 reference dates on the x-axis

Inverting the x-axis since the csv file starts from the current date
